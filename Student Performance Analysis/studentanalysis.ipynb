```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
print("Dataset shape:", df.shape)
print("Columns:", df.columns.tolist())
```

```python
print(df.info())
print(df.describe())
print("Missing values:\n", df.isnull().sum())
```

```python
# Gender distribution
plt.figure(figsize=(6,4))
sns.countplot(x='gender', data=df, palette='pastel')
plt.title('Gender Distribution')
plt.show()

# Parental education level
plt.figure(figsize=(10,5))
sns.countplot(x='parental level of education', data=df, palette='muted', order=df['parental level of education'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Parental Education Level Distribution')
plt.show()
```

```python
plt.figure(figsize=(12, 8))

# Math Score Distribution
plt.subplot(3, 1, 1)
sns.histplot(df['math score'], kde=True, color='skyblue', bins=15)
plt.title('Math Score Distribution')

# Reading Score Distribution
plt.subplot(3, 1, 2)
sns.histplot(df['reading score'], kde=True, color='orange', bins=15)
plt.title('Reading Score Distribution')

# Writing Score Distribution
plt.subplot(3, 1, 3)
sns.histplot(df['writing score'], kde=True, color='green', bins=15)
plt.title('Writing Score Distribution')

plt.tight_layout()
plt.show()
```

```python
score_cols = ['math score', 'reading score', 'writing score']
plt.figure(figsize=(6,4))
corr = df[score_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Score Correlations')
plt.show()
```
